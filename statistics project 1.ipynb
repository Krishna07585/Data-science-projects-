# Standardization of Numerical Variables
●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
●	Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
 
 Show before and after comparisons of the data distributions.


```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Show before standardization
print("Before Standardization:")
print(df[numerical_columns].describe())

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the numerical columns
df_standardized = df.copy()
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Plot histograms before and after standardization

plt.figure(figsize=(20, 15))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, 2*i-1)
    sbs.histplot(df[column], kde=True, bins=30)
    plt.title(f'Before Standardization: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    plt.subplot(len(numerical_columns), 2, 2*i)
    sbs.histplot(df_standardized[column], kde=True, bins=30)
    plt.title(f'After Standardization: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
